```python
from sklearn.preprocessing import LabelEncoder as le
```

```python
df["gender"] = le().fit_transform(df["gender"])
```

```python
from sklearn.preprocessing import OrdinalEncoder as oe

df["customer_level"] = oe(categories=[["Bronze", "Silver", "Gold", "Platinum"]]).fit_transform(df[["customer_level"]])
```

```python
from sklearn.preprocessing import OneHotEncoder 

ohe = OneHotEncoder(sparse_output=False)

encoded = ohe.fit_transform(df[["preferred_device"]])
```

```python
encoded_df = pd.DataFrame(
    encoded,
    columns = ohe.get_feature_names_out(["preferred_device"])
)
```

```python
df = pd.concat([df,encoded_df],axis=1)
```

```python
df.head(10)
```

```python
from sklearn.preprocessing import StandardScaler 

scaler = StandardScaler()

df[["age", "income"]] = scaler.fit_transform(
    df[["age", "income"]]
)
```

```python
from sklearn.preprocessing import MinMaxScaler

mmscale = MinMaxScaler()

df[["satisfaction_score"]] = mmscale.fit_transform(
    df[["satisfaction_score"]]
)
```

```python
df.head(10)
```

```python
from sklearn.impute import KNNImputer, SimpleImputer

knn = KNNImputer(n_neighbors=2)

df = pd.DataFrame(
    knn.fit_transform(df),
    columns=df.columns
)
```
