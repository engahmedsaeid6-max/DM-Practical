```python
import random

# Items: [weight, value]
items = [
    [1, 2],
    [2, 4],
    [3, 4],
    [4, 5],
    [5, 7],
    [6, 9]
]

max_weight = 10
population_size = 10
mutation_probability = 0.2
generations = 10


# --------- Individual ----------
def create_individual():
    return [random.randint(0, 1) for _ in range(len(items))]


# --------- Fitness ----------
def fitness(individual):
    weight = 0
    value = 0

    for i in range(len(items)):
        if individual[i] == 1:
            weight += items[i][0]
            value += items[i][1]

    return value if weight <= max_weight else 0


# --------- Selection ----------
def select_parent(population):
    return max(random.sample(population, 3), key=fitness)


# --------- Crossover ----------
def crossover(p1, p2):
    point = random.randint(1, len(items) - 1)
    return p1[:point] + p2[point:]


# --------- Mutation ----------
def mutate(ind):
    for i in range(len(ind)):
        if random.random() < mutation_probability:
            ind[i] = 1 - ind[i]
    return ind


# --------- MAIN GENETIC ALGORITHM ----------
def genetic_algorithm():
    population = [create_individual() for _ in range(population_size)]

    best_solution = None
    best_fitness = 0

    for gen in range(generations):

        new_population = []

        for _ in range(population_size // 2):

            parent1 = select_parent(population)
            parent2 = select_parent(population)

            child1 = mutate(crossover(parent1, parent2))
            child2 = mutate(crossover(parent2, parent1))

            new_population.extend([child1, child2])

        population = new_population

        # track best
        for ind in population:
            f = fitness(ind)
            if f > best_fitness:
                best_fitness = f
                best_solution = ind

        print(f"Generation {gen+1}: Best = {best_fitness}")

    return best_solution, best_fitness


# --------- RUN ----------
best, value = genetic_algorithm()

print("\nBest solution:", best)
print("Best value:", value)
```
